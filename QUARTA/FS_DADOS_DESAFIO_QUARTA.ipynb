##Importando bibliotecas

```python
#renomeando as bibliotecas para melhor praticidade

import pandas as pd
import matplotlib.pyplot as plt
```

##Carregando dados

```python
#acessando a biblioteca com o pd
#lendo os dados do arquivo baixado com o comando read

df = pd.read_csv('/content/dados.csv')
df
```

##Pandas

```python
#operaÃ§Ãµes com os campos matemÃ¡tica, leitura e escrita

#crio uma variÃ¡vel, acesso o arquivo com o dataframe 'df', nome da coluna
#quantidade de casas(nÃ£o tem, pois quero calcular tudo)
#por fim os comandos especÃ­ficos das operaÃ§Ãµes

#calculando valores mÃ­nimos

min_matematica = df['math score'].min()
min_leitura = df['reading score'].min()
min_escrita = df['writing score'].min()

#calculando valores mÃ¡ximos

max_matematica = df['math score'].max()
max_leitura = df['reading score'].max()
max_escrita = df['writing score'].max()

#calculando valore mÃ©dios

med_matematica = df['math score'].mean()
med_leitura = df['reading score'].mean()
med_escrita = df['writing score'].mean()

#apresentando os resultados

print("[MatemÃ¡tica] mÃ­nimo:", min_matematica, ", mÃ¡ximo:", max_matematica, ", mÃ©dia:", med_matematica)
print("[Leitura] mÃ­nimo:", min_leitura, ", mÃ¡ximo:", max_leitura, ", mÃ©dia:", med_leitura)
print("[Escrita] mÃ­nimo:", min_escrita, ", mÃ¡ximo:", max_escrita, ", mÃ©dia:", med_escrita)
```

##Pandas + Matplotlib

```python
#banco de dados a partir da raÃ§a/etnia dos alunos

#nota mÃ©dia em matemÃ¡tica para cada grupo disponÃ­vel

#um array dentro de vÃ¡rios arrays, array multidimensional, acesso ele por meio dos colchetes
#crio um novo dataframe para nÃ£o perder as informaÃ§Ãµes do antigo ou sobrescrevÃª-lo

df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2
```

```python
#grÃ¡fico do df2

#estilizando o grÃ¡fico
#modelo, color, tÃ­tulo e nomes dos eixos para melhor visualizaÃ§Ã£o

df2.plot(kind='barh', color ='#4305af')
plt.title('Nota mÃ©dia em matemÃ¡tica por raÃ§a/etnia')
plt.ylabel('RaÃ§a/etnia')
plt.xlabel('Nota mÃ©dia em matemÃ¡tica')
```

```python
#banco de dados a partir do nÃ­vel de escolaridade dos pais

#nota mÃ©dia em escrita para cada grupo disponÃ­vel

df3 = df.groupby('parental level of education')[['writing score']].mean()
df3
```

```python
#grÃ¡fico do df3

#estilizando o grÃ¡fico
#modelo, color, tÃ­tulo e nomes dos eixos para melhor visualizaÃ§Ã£o

df3.plot(kind='barh', color ='#459efd')
plt.title('Nota mÃ©dia em escrita por nÃ­vel de escolaridade dos pais')
plt.ylabel('NÃ­vel de escolaridade dos pais')
plt.xlabel('Nota mÃ©dia em escrita')
```

```python
#banco de dados a partir da coluna referente Ã  participaÃ§Ã£o em um curso preparatÃ³rio para os testes

#nota mÃ©dia em leitura para cada grupo disponÃ­vel

df4 = df.groupby('test preparation course')[['reading score']].mean()
df4
```

```python
#grÃ¡fico do df4

#estilizando o grÃ¡fico
#modelo, color, tÃ­tulo e nomes dos eixos para melhor visualizaÃ§Ã£o

df4.plot(kind='barh', color ='#45e4fd')
plt.title('Nota mÃ©dia em leitura por participaÃ§Ã£o em um curso preparatÃ³rio')
plt.ylabel('ParticipaÃ§Ã£o em um curso preparatÃ³rio')
plt.xlabel('Nota mÃ©dia em leitura')
```
