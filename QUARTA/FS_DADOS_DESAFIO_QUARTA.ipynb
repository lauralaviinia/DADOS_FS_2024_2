##Importando bibliotecas

```python
#renomeando as bibliotecas para melhor praticidade

import pandas as pd
import matplotlib.pyplot as plt
```

##Carregando dados

```python
#acessando a biblioteca com o pd
#lendo os dados do arquivo baixado com o comando read

df = pd.read_csv('/content/dados.csv')
df
```

##Pandas

```python
#operações com os campos matemática, leitura e escrita

#crio uma variável, acesso o arquivo com o dataframe 'df', nome da coluna
#quantidade de casas(não tem, pois quero calcular tudo)
#por fim os comandos específicos das operações

#calculando valores mínimos

min_matematica = df['math score'].min()
min_leitura = df['reading score'].min()
min_escrita = df['writing score'].min()

#calculando valores máximos

max_matematica = df['math score'].max()
max_leitura = df['reading score'].max()
max_escrita = df['writing score'].max()

#calculando valore médios

med_matematica = df['math score'].mean()
med_leitura = df['reading score'].mean()
med_escrita = df['writing score'].mean()

#apresentando os resultados

print("[Matemática] mínimo:", min_matematica, ", máximo:", max_matematica, ", média:", med_matematica)
print("[Leitura] mínimo:", min_leitura, ", máximo:", max_leitura, ", média:", med_leitura)
print("[Escrita] mínimo:", min_escrita, ", máximo:", max_escrita, ", média:", med_escrita)
```

##Pandas + Matplotlib

```python
#banco de dados a partir da raça/etnia dos alunos

#nota média em matemática para cada grupo disponível

#um array dentro de vários arrays, array multidimensional, acesso ele por meio dos colchetes
#crio um novo dataframe para não perder as informações do antigo ou sobrescrevê-lo

df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2
```

```python
#gráfico do df2

#estilizando o gráfico
#modelo, color, título e nomes dos eixos para melhor visualização

df2.plot(kind='barh', color ='#4305af')
plt.title('Nota média em matemática por raça/etnia')
plt.ylabel('Raça/etnia')
plt.xlabel('Nota média em matemática')
```

```python
#banco de dados a partir do nível de escolaridade dos pais

#nota média em escrita para cada grupo disponível

df3 = df.groupby('parental level of education')[['writing score']].mean()
df3
```

```python
#gráfico do df3

#estilizando o gráfico
#modelo, color, título e nomes dos eixos para melhor visualização

df3.plot(kind='barh', color ='#459efd')
plt.title('Nota média em escrita por nível de escolaridade dos pais')
plt.ylabel('Nível de escolaridade dos pais')
plt.xlabel('Nota média em escrita')
```

```python
#banco de dados a partir da coluna referente à participação em um curso preparatório para os testes

#nota média em leitura para cada grupo disponível

df4 = df.groupby('test preparation course')[['reading score']].mean()
df4
```

```python
#gráfico do df4

#estilizando o gráfico
#modelo, color, título e nomes dos eixos para melhor visualização

df4.plot(kind='barh', color ='#45e4fd')
plt.title('Nota média em leitura por participação em um curso preparatório')
plt.ylabel('Participação em um curso preparatório')
plt.xlabel('Nota média em leitura')
```

###**análise final dos gráficos**

**gráfico 1**: banco de dados a partir da raça/etnia dos alunos calculando a nota média em matemática para cada grupo disponível.

o "grupo E" tem tendência a ter a média de notas maiores que todos os outros, já o "grupo A" tende a notas menores.

**gráfico 2**: banco de dados a partir do nível de escolaridade dos pais calculando a nota média em escrita para cada grupo disponível.

ter mestrado tende a mostrar uma média maior que os demais enquanto ter apenas o ensino médio tende a possuir menor média.

**gráfico 3**: banco de dados a partir da coluna referente à participação em um curso preparatório para os testes calculando a nota média em leitura para cada grupo disponível.

os que fazem curso preparatório tem a média maior daqueles que não fazem.
